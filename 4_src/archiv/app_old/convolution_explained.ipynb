```python
import numpy as np
import skimage
from skimage import io
import pandas as pd
import matplotlib.pyplot as plt
```

```python
patch = io.imread('./41700_41700_0_19.png', as_gray=True)
```

```python
io.imshow(patch)
plt.show()
```

```python
smooth_image = skimage.filters.gaussian(patch, 5)
```

```python
io.imshow(smooth_image)
plt.show()
```

```python
horizontal_edge_kernel = np.array([[ 1,  2,  1],
                                   [ 0,  0,  0],
                                   [-1, -2, -1]])


plt.figure(figsize=(20,20))
plt.imshow(horizontal_edge_kernel, cmap=plt.cm.RdBu);
plt.show()
```

```python
from scipy.ndimage import convolve


plt.figure(figsize=(20,20))
horizontal_edge_response = convolve(smooth_image, horizontal_edge_kernel)
plt.imshow(horizontal_edge_response, cmap=plt.cm.RdBu);
plt.show()
```
