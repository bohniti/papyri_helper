```python
import numpy as np
import skimage
from skimage import io
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

%matplotlib inline
```

```python
from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate sample data
n_samples = 4000
n_components = 4

X, y_true = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=2, random_state=0
)
X = X[:, ::-1]

# Calculate seeds from kmeans++
centers_init, indices = kmeans_plusplus(X, n_clusters=4, random_state=0)

# Plot init seeds along side sample data
fig =plt.figure(figsize=(25,15))
fig.patch.set_facecolor('#333333')
colors = ["#80cdc1", "#a6611a", "#018571", "#dfc27d"]

ax = fig.add_subplot(111)
ax.patch.set_facecolor('#333333')

for k, col in enumerate(colors):
    cluster_data = y_true == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1], c=col, marker=".", s=200)

#plt.scatter(centers_init[:, 0], centers_init[:, 1], c="b", s=50)
plt.xticks([])
plt.yticks([])
plt.savefig('./embeddings_1.png')
plt.show()
```

```python
from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
n_samples = 2000
n_components = 2

X_1, y_true_1 = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=0.3, random_state=0
)
X_1 = X_1[:, ::-1]
```

```python
n_samples = 2000
n_components = 2

X_2, y_true_2 = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=0.3, random_state=0
)
X_2 = X_2[:, ::-1]

y_true_2 = np.where(y_true_2 ==0,2, y_true_2)
y_true_2 = np.where(y_true_2 ==1,3, y_true_2)
```

```python
X = np.concatenate((X_1-0.8, X_2))
```
